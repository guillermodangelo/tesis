```python
import geopandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lines = geopandas.read_file("capas/mig_recientes_2011.gpkg", layer='flujos')
lines.head()
```

```python
lines_no_mvo = lines.loc[(lines.depto_origen > 1) & (lines.depto_destino > 1)]
```

```python
lines_no_mvo.head()
```

```python
# genera tabla pivot con los flujos de departamento a departamento
matrix = pd.DataFrame(pd.pivot_table(lines,
                                    index ='depto_origen',
                                    columns='depto_destino',
                                    fill_value=0,
                                    aggfunc=sum,
                                    margins=False))

# genera tabla pivot con los flujos de departamento a departamento
matrix_no_mvo = pd.DataFrame(pd.pivot_table(lines_no_mvo,
                                    index ='depto_origen',
                                    columns='depto_destino',
                                    fill_value=0,
                                    aggfunc=sum,
                                    margins=False))

matrix_no_mvo
```

```python
# el máximo número de migrantes sin contar Mvdeo. como origen/destino es con origen en Canelones y destino Maldonado.
lines_no_mvo.loc[lines_no_mvo.personas_mig == max(lines_no_mvo.personas_mig)]
```

```python
array = matrix.values
array_no_mvo = matrix_no_mvo.values
```

```python
np.amax(array_no_mvo)
```

```python
nomdep = ['Montevideo', 'Artigas', 'Canelones', 'Cerro Largo', 'Colonia', 'Durazno',
          'Flores', 'Florida', 'Lavalleja','Maldonado', 'Paysandú', 'Río Negro', 'Rivera',
          'Rocha', 'Salto', 'San José', 'Soriano', 'Tacuarembó', 'Treinta y Tres']
```

```python
# a 2D array with linearly increasing values on the diagonal
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111)
cax = ax.matshow(array, cmap='Reds_r')

plt.xticks(np.arange(0, 19, step=1), labels=nomdep, rotation=90)
plt.yticks(np.arange(0, 19, step=1), labels=nomdep)

ax.tick_params(axis=u'both', which=u'both',length=0)

plt.show()
```

```python
# a 2D array with linearly increasing values on the diagonal
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111)
cax = ax.matshow(array_no_mvo, cmap='Reds_r')

plt.xticks(np.arange(0, 18, step=1), labels=nomdep[1:19], rotation=90)
plt.yticks(np.arange(0, 18, step=1), labels=nomdep[1:19])

ax.tick_params(axis=u'both', which=u'both',length=0)

plt.show()
```
