```python
lines = geopandas.read_file("capas/mig_recientes_2011.gpkg", layer='flujos')
```

```python
for i in range(1,20):
    lines_filtered = lines.loc[lines.depto_origen== i]
    file_name = "capas/geojson/mig_recientes_2011_depori_{}.geojson".format(i)
    lines_filtered.to_file(file_name, driver='GeoJSON')
```

```python
lines_json = lines.loc[lines.depto_origen== 1].to_crs(epsg='4326').to_json()
```

```python
lines_json = [lines.loc[lines.depto_origen==x].to_crs(epsg='4326').to_json() for x in range(1,20)]
```

```python
lines_json[0]
```

```python
names = [
'Montevideo',
'Artigas',
'Canelones',
'Cerro Largo',
'Colonia',
'Durazno',
'Flores',
'Florida',
'Lavalleja',
'Maldonado',
'Paysandú',
'Río Negro',
'Rivera',
'Rocha',
'Salto',
'San José',
'Soriano',
'Tacuarembó',
'Treinta y Tres'
]

names[18]
```

```python
mapa = folium.Map([-32, -56],
                  zoom_start=7)

folium.features.GeoJson(data=lines_json[0],
                                  tooltip=folium.features.GeoJsonTooltip(
                                        fields=['personas_mig'],
                                        aliases=['Migrantes:'], 
                                        labels=True, 
                                        sticky=False),
                                  name='Mvdeo.').add_to(mapa)
    
for i in range(1,19):
    folium.features.GeoJson(data=lines_json[i],
                                  tooltip=folium.features.GeoJsonTooltip(
                                        fields=['personas_mig'],
                                        aliases=['Migrantes:'], 
                                        labels=True, 
                                        sticky=False),
                                  name=names[i],
                           show=False).add_to(mapa)

# https://python-visualization.github.io/folium/modules.html#folium.features.GeoJson
# mvo = folium.features.GeoJson(lines_json[1], name='Artigas', show=False).add_to(mapa)

# folium.GeoJson(
#     lines_json,
#     tooltip=folium.features.GeoJsonTooltip(fields=['feature.properties.personas_mig'])
# ).add_to(mvo)



folium.LayerControl(collapsed=False).add_to(mapa) 

mapa

# mapa.save('prueba_1.html')
```
